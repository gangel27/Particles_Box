```python
import random

def step(state):
    a, b, c = state
    possible_moves = []

    # Define all valid collisions
    if a > 0 and b > 0:
        possible_moves.append((a-1, b-1, c+2))  # A + B -> C + C
    if b > 0 and c > 0:
        possible_moves.append((a+2, b-1, c-1))  # B + C -> A + A
    if c > 0 and a > 0:
        possible_moves.append((a-1, b+2, c-1))  # C + A -> B + B

    if not possible_moves:
        return state  # no move possible

    return random.choice(possible_moves)


def is_equilibrium(state):
    a, b, c = state
    return (a == 0 and b == 0) or (b == 0 and c == 0) or (c == 0 and a == 0)


def simulate(initial_state, max_steps=100):
    state = initial_state
    seen = set()

    for _ in range(max_steps):
        if state in seen:
            return "cycle"
        seen.add(state)

        if is_equilibrium(state):
            return "equilibrium"

        state = step(state)

    return "unknown"  # in case it takes too long


def test_range(n):
    results = {}

    for a in range(n+1):
        for b in range(n+1 - a):
            c = n - a - b
            state = (a, b, c)

            outcome = classify_state(state)
            results[state] = outcome

    return results

def classify_state(state, trials=200):
    outcomes = set()
    for _ in range(trials):
        outcomes.add(simulate(state))
    return outcomes


# Example usage:
results = test_range(100)

for state, outcome in results.items():
    print(f"{state} = {(state[1] - state[2]) % 3, (state[0] - state[2]) % 3, (state[0] - state[1]) % 3} -> {outcome}")
```

```python

def next_states(state):
    a, b, c = state
    moves = []

    if a > 0 and b > 0:
        moves.append((a-1, b-1, c+2))
    if b > 0 and c > 0:
        moves.append((a+2, b-1, c-1))
    if c > 0 and a > 0:
        moves.append((a-1, b+2, c-1))

    return moves


def equilibrium_label(state):
    a, b, c = state
    if b == 0 and c == 0:
        return "A"
    if a == 0 and c == 0:
        return "B"
    if a == 0 and b == 0:
        return "C"
    return None


def classify_exact(state, memo=None, path=None):
    if memo is None:
        memo = {}
    if path is None:
        path = set()

    # Already computed
    if state in memo:
        return memo[state]

    # Cycle detection
    if state in path:
        return {"cycle"}

    # Equilibrium
    label = equilibrium_label(state)
    if label:
        memo[state] = {label}
        return {label}

    path.add(state)

    outcomes = set()

    for nxt in next_states(state):
        outcomes |= classify_exact(nxt, memo, path)

    path.remove(state)

    memo[state] = outcomes
    return outcomes

eg = (0, 8, 92) 
print(classify_exact((85,3,12)))  
print(classify_exact(eg))
```
